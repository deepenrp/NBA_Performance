```python
import pandas as pd
```

```python
fdp= pd.read_csv('fdp.csv')
steals=pd.read_csv('steals_prediction.csv')
assists=pd.read_csv('Assist.csv')
blocks=pd.read_csv('blocks.csv')
turnovers=pd.read_csv('turnover.csv')
rebound=pd.read_csv('rebound.csv')
points=pd.read_csv('pnt_predict.csv')
```

```python
steals=steals[["Player_name","ST"]]
assists=assists[["Player_name","AS"]]
blocks=blocks[["Player_name","BL"]]
turnovers=turnovers[["Player_name","TO"]]
rebound=rebound[["Player_name","RB"]]
points=points[["Player_name","PT"]]
```

```python
merge_df= fdp.merge(steals,  on= 'Player_name', how = 'inner')
merge1= merge_df.merge(assists,  on= 'Player_name', how = 'inner')
merge2=merge1.merge(blocks, on= 'Player_name', how = 'inner')
merge3=merge2.merge(turnovers, on= 'Player_name', how = 'inner')
merge4=merge3.merge(rebound, on= 'Player_name', how = 'inner')
merge5=merge4.merge(points, on= 'Player_name', how = 'inner')
```

```python
merge5
```

```python
Final_data= merge5[['FDP_pd', 'Player_name', 'ID', 'Date', 'Team', 'Opp', 'H/A',
       'GTime(ET)', 'Team pts', 'Opp pts', 'Win-Lose', 'Start', 'Mins_played',
       'GP', 'active', 'DoubleD', 'TripleD', 'FD Sal', 'FD pos', 'ADI',
       'VMI', 'trey', 'FGM', 'FGA', 'FTM',
       'FTA', 'ranking', 'ST', 'AS', 'BL', 'TO', 'RB', 'PT']]
```

```python
# Final_data.rename(columns = {0:‘Perdictions’}, inplace = True)
Final_data.rename(columns={'FDP_pd': 'FDP'}, inplace=True)
```
